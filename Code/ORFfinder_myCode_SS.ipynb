```python
# Import the SeqIO module from Biopython and read the fasta file into the 'record' object
record = SeqIO.read("/content/drive/MyDrive/Life Sciences/phage_MS2_genome_sequence.fasta", "fasta")

# Extract the DNA sequence from the 'record' object and assign it to the 'input_seq' variable
input_seq = record.seq

# Define the start codon sequence as "ATG" and the stop codon sequences as "TGA", "TAG", and "TAA"
start_codons_seq = "ATG"
stop_codons_seqs = ["TGA","TAG","TAA"]

# Initialize variables for the ORF number, frame number, ORF begin position, ORF end position, and ORF sequences
ORF_number = 0
frame_number = 1
ORF_begin = 0
ORF_end = 0
ORF_seqs = dict()

# Define the length of ORFs to be 30 nucleotides
length_value = 30

#defining start postion of the necluitide
nuc_pos = 0

# Loop through each nucleotide in the input DNA sequence, checking for start and stop codons
# while loop will search through the entire DNA sequence, one codon at a time, until the end of the sequence is reached.
while nuc_pos < (len(input_seq) - 2):
  # This line extracts a three-nucleotide codon from the DNA sequence starting at the current position and assigns it to the 'codon' variable.
  codon = (input_seq[nuc_pos:(nuc_pos + 3)])
  # If a start codon is found, loop through the sequence from that point on, checking for a stop codon
  if codon == start_codons_seq:
    # This line starts a for loop that will search through the DNA sequence starting at the position of the start codon, looking for the next stop codon.
    for nuc_pos_2 in range(nuc_pos + 3,(len(input_seq) - 2),3):
      # This line extracts a three-nucleotide codon from the DNA sequence starting at the current position in the for loop and assigns it to the 'codon1' variable.
      codon1 = (input_seq[nuc_pos_2:(nuc_pos_2 + 3)])
      # This line checks if the 'codon1' variable matches any of the stop codon sequences specified earlier. If it does, the code will check if the length of the ORF is greater than the minimum length specified earlier. If it is, the ORF is considered a valid hit and its details are recorded.
      if codon1 in stop_codons_seqs:
        # If a stop codon is found, and the length of the ORF is greater than 30 nucleotides, add it to the ORF_seqs dictionary
        # These lines increase the ORF count, assign the beginning and end positions of the ORF, store the ORF sequence in the 'ORF_seqs' dictionary, and move the search position to the end of the ORF so that the code will not search within it again.
        if (nuc_pos_2 - nuc_pos) > length_value:
          ORF_number = ORF_number + 1
          ORF_begin = nuc_pos
          ORF_end = nuc_pos_2 + 2
          ORF_seqs["ORF" + str(ORF_number)] = input_seq[nuc_pos:nuc_pos_2 + 3]
          nuc_pos = nuc_pos_2
          # This line exits the for loop once a stop codon is found.
        break
        # Continue searching for ORFs until the end of the sequence is reached
        # This line increments the search position by three nucleotides to move to the next codon.
  nuc_pos = nuc_pos + 3
    
# Print the number of ORFs, ORF begin position, ORF end position, and ORF sequences
print(ORF_number)
print(ORF_begin)
print(ORF_end)
print(ORF_seqs)
```
