```python
# defining a function for calculating GC percent
def calcGC(my_seq):
  # counting every neculiotide
  g_count = my_seq.count('G')
  c_count = my_seq.count('C')
  # formula for counting GC% in the given sequence
  GC_per = 100 * (g_count + c_count)/ len(my_seq)
  return(GC_per) 
```

```python
# Set input sequence as a string
input_sequence = "ATAAGACTCCTTAAGTTCTCTACAATGTACGACCCGGAAAGGTTGCACAACTTAGACCCTACAATGTACTAGTAGGCCTCTGCGCTGGCAATACAGATAAGATAATGTAGCCCCTGGCCTCAAAGGAACTCTCCTCCTTAGGTTGCATTTGTATAATGTTTGATTTTTAGAACAAAGGTTGCATGTACTAGTTCTCCTCCTTAAGTTCTCGTTCTCCTCCAATAAGAATAAGATAATGTT"

# Set start and stop regions of the forward primer and the maximum buffer
start_region = 29
stop_region = 79
max_buffer = 10

# Set primer length
primer_length = 20

# For loop to iterate through primer start positions
for primer_start in range(start_region, (start_region - max_buffer), -1): 

  # Read input sequence from primer_start to primer_start+primer_length
  read_sequence = (input_sequence[primer_start:(primer_start + primer_length)])
  
  # Calculate the ratio of GC to AT nucleotides in the read sequence
  if 45 < calcGC(read_sequence) < 55:
    # Set the output sequence to be the read sequence
    output_seq = read_sequence
    
    # Print the GC ratio and the output sequence
    print(calcGC(read_sequence))
    print(output_seq)
```
